```python
# Introduza aqui funcoes gerais
import numpy as np
import numpy.random as nr
import matplotlib.pyplot as plt
import scipy as sp
from scipy.special import factorial as fac
from matplotlib.animation import FuncAnimation
plt.rcParams['figure.figsize'] = [20 , 10]
SMALL_SIZE, MEDIUM_SIZE, BIGGER_SIZE  = 20, 24, 30
plt.rc('font', size=SMALL_SIZE)          
plt.rc('axes', titlesize=BIGGER_SIZE)
plt.rc('axes', labelsize=MEDIUM_SIZE)    
plt.rc('xtick', labelsize=SMALL_SIZE)
plt.rc('ytick', labelsize=SMALL_SIZE)   
plt.rc('legend', fontsize=SMALL_SIZE)    
plt.rc('figure', titlesize=BIGGER_SIZE)  

class ising:
    def __init__(self,N,betas):
        self.N = N
        self.beta = betas
        self.M = np.arange(-N, N + 1, 2)
        self.E = (N - self.M) * (N + self.M) / (2 * (N - 1))
        self.A = (N + self.M) / 2
        self.B = (N - self.M) / 2
        self.LO = self.lnfac(N) - self.lnfac(self.A) - self.lnfac(self.B)
        
        self.temp0 = self.LO - self.beta[:,None]*self.E
        self.C = np.amax( self.temp0 ,axis=1)
        self.temp = self.temp0 - self.C[:,None]
        self.Z = np.sum(np.exp(self.temp),axis=1)
        self.p = np.exp(self.temp) / self.Z[:,None]
        self.mag = np.sum((np.abs(self.M)/N) * self.p,axis=1)
        
    def lnfac(self,A):
        app = 20
        if type(A) != int:
            B = np.empty(A.size)
            i1 , i2 = np.where(A <= app), np.where(A > app)
            B[i1] = np.log(fac(A[i1]))
            B[i2] = A[i2] * (np.log(A[i2])-1) + np.log(2*np.pi*A[i2])/2
            return B
        else:
            if A <= app:
                return np.log(fac(A))
            else:
                return A * (np.log(A)-1) + np.log(2*np.pi*A)
```

# Exercício 1
Nas aulas discutimos a solução do modelo de Ising de campo médio com N spins. Neste trabalho de casa vamos revisitar-lo numericamente. Considere o Hamiltoniano, 
\begin{equation}
H=\frac{1}{2}\sum_{i=1}^{N}\left(1-s_{i}\frac{M-s_{i}}{N-1}\right)=\frac{(N-M)(N+M)}{2(N-1)}
\end{equation}
onde a magnetização é dada por $M=\sum_{i=1}^{N}s_{i}$ e as variáveis de spin $s_{i}$ são $\{\pm1\}$.  Para este modelo a energia é função apenas da magnetização e do número de partículas. A densidade de estados do sistema é obtida usando o número de spins $s_{i}=+1$ ($N_{+}$) e o número de spins $s_{i}=-1$ ($N_{-}$): 
\begin{align}
\Omega(N_{+},N_{-})	&=\frac{N!}{N_{+}!N_{-}!}& 	N_{+}	&=\frac{N+M}{2}	& N_{-}	&=\frac{N-M}{2}
\end{align}
Com esta parametrização,
\begin{equation}
\Omega(M)=\frac{N!}{\left(\frac{N+M}{2}\right)!\left(\frac{N-M}{2}\right)!}
\end{equation}
onde $M\in\{-N,-N+2,\dots,N-2,N\}$.

1. Calcule numericamente o valor médio exacto como função da temperatura do módulo da magnetização para $\beta$ entre 0.1 e 1.5.
\begin{align}
\left\langle \frac{|M|}{N}\right\rangle 	& =\sum_{M}\frac{|M|}{N}\Omega(M)\frac{e^{-\beta E(M)}}{Z(\beta)}  \\
Z(\beta)	& =\sum_{M}\Omega(M)e^{-\beta E(M)}
\end{align}
**Sugestão**: O Cálculo da combinatórica envolve números muito grandes. O idela é usar a fórmula de Stirling,
\begin{equation}
\log(n!)=n\log(n)-n+\frac{1}{2}\log(2\pi n)
\end{equation}
e notar que o valor médio fica inalterado se deslocar a energia $-\beta E(M)\to-\beta E(M)-C$ nas exponenciais. Logo o ideal é calcular,
$$\left\langle \frac{|M|}{N}\right\rangle =\frac{\sum_{M}\frac{|M|}{N}\Omega(M)e^{-\beta E(M)-C}}{\sum_{M}\Omega(M)e^{-\beta E(M)-C}}$$
onde C é o máximo de $\ln\Omega(M)-\beta E(M)$ para cada temperatura.

(b) Meça a autocorrelação da magnetização para $\beta=1.05$,
\begin{align}
corr_{M}(\tau)	& =\frac{\left\langle M_{i}M_{i+\tau}\right\rangle -\left\langle M_{i}\right\rangle ^{2}}{\left\langle M_{i}^{2}\right\rangle -\left\langle M_{i}\right\rangle ^{2}} \\
\left\langle M_{i}M_{i+\tau}\right\rangle 	& =\frac{1}{L}\sum_{i=1}^{L}M_{i}M_{i+\tau}
\end{align}


 (#) Pode  Escrever aqui texto // Como se pode observar, a correlação é desprezável e não apresenta qualquer dependência do tamanho da amostra, como seria de esperar já que todos os eventos da amostra são completamente independentes.

```python
# Escreva aqui codigo
def corre(amostra,corte):
    fim = amostra.size
    Mi = amostra[:corte]
    med = np.mean(amostra)
    var = np.mean(amostra**2) - med**2
    tau = np.arange(1,int(fim/10),5)
    CORR = np.empty(tau.size)
    for i in range(CORR.size):
        CORR[i] = (( np.mean(Mi*amostra[tau[i]:corte+tau[i]]) ) - med**2)/var
    return tau,CORR

amostra_direto = nr.choice(two.M,2**18,p=two.p[5])

tau, corr = corre(amostra_direto,1000)
plt.plot(tau,corr), plt.title('Correlação'), plt.xlabel(r'$\tau$'), plt.ylabel('Correlação')
plt.show()
```

3.  Que implicações tem a autocorrelação nos erros das medidas?

 (#) Pode  Escrever aqui texto //
A variância do estimador do valor médio é dada por:
$$ \left\langle\bar{X}^{2}\right\rangle-\langle\bar{X}\rangle^{2}=\frac{1}{N^{2}} \sum_{i j}\left\langle x_{i} x_{j}\right\rangle-\langle x\rangle^{2} $$

$$ =\frac{1}{N^{2}} \sum_{i}\left\langle x_{i}^{2}\right\rangle+\frac{1}{N^{2}} \sum_{i \neq j}\left\langle x_{i} x_{j}\right\rangle-\langle x\rangle^{2}$$

No caso de todos os elementos da amostra serem independentes obtém-se:

$$\left\langle\bar{X}^{2}\right\rangle-\langle\bar{X}\rangle^{2}=\frac{\left\langle x^{2}\right\rangle}{N}+\frac{N-1}{N}\langle x\rangle^{2}-\langle x\rangle^{2}$$

$$=\frac{\left\langle x^{2}\right\rangle-\langle x\rangle^{2}}{N}$$

$$var(\bar{X}) = \frac{var({X})}{N}$$

Se se obtiver uma amostra através do algoritmo de Metropolis construído, os seus elementos não são independentes, já que a probabilidade de aceitar uma sugestão de um novo elemento para a amostra depende do elemento anterior. Neste caso obtém-se uma expressão da forma:

$$var(\bar{X}) = \frac{var({X})}{N} (1+C(\zeta))$$

Onde $C$ é uma função de $\zeta$, que é o integral da correlação até o $\tau$, o tamanho da amostra.

A partir de um certo $\tau'$, o tamanho de "termalização", a correlação é muito próxima de 0, ou oscila em torno deste, e então o seu integral é aproximadamente constante. Como $C(\zeta)$ é constante, a variância do estimador de valor médio é proporcional ao inverso de $M$. No entanto, antes desse $\tau'$, $C(\zeta)$ pode ter uma dependência relevante com M, por exemplo linear, com a qual var(\bar{X}) decai muito menos rapidamente, podendo ser aproximadamente constante.

Deste modo, ao calcular o valor do estimador do valor médio, é importante que o tamanho da amostra seja maior que $\tau'$, já que só nesse caso a variância desse valor é desprezável. Uma média estimada com uma amostra de tamanho menor que $\tau'$ seria estatisticamente equivalente a retirar uma só medição; variar o tamanho da amostra dentro desse regime é essencialmente irrelevante para a média estimada.

# Exercício 3
Na pergunta anterior efectuamos uma simulação Monte Carlo num espaço de fase a uma dimensão, da variável M. Será que esta dinâmica corresponde à mesma dinâmica do sistema de N spins com o Hamiltoniano 
$$H=\frac{1}{2}\sum_{i=1}^{N}\left(1-s_{i}\frac{M-s_{i}}{N-1}\right)$$ de campo médio e o algorítmo de Metropolis?

1. Para testar implemente o algoritmo de Metropolis para $N$ spins:

• Atribua a cada um dos N spins do sistema uma variável \pm1 com probailidade uniforme.

• Considere um passo de tempo o seguinte algoritmo:

– Escolha um spin com probabilidade uniforme.

– Aceite inverter-lo com a probabilidade $\min\left(1,e^{-\beta\Delta E}\right)$ senão fique no mesmo estado.


 (#) Pode  Escrever aqui texto 

```python
# Escreva aqui codigo
def H(spins):
    N, M = spins.size, np.sum(spins)
    return (N-M)*(N+M)/(2*(N-1))

def carlo(N,tam,beta):
    j = nr.choice([-1, 1], N)
    amostra = np.empty(tam)
    for t in range(2**10):
        ind = nr.randint(N)
        i = np.copy(j)
        i[ind] = j[ind] * -1
        if nr.rand() < min(1,np.exp(-beta*(H(i)-H(j)))):
            j = np.copy(i)
    for t in range(tam):
        ind = nr.randint(N)
        i = np.copy(j)
        i[ind] = j[ind] * -1
        if nr.rand() < min(1,np.exp(-beta*(H(i)-H(j)))):
            j = np.copy(i)
        amostra[t] = np.sum(j)
    return amostra
N = 55
last = ising(N,np.array([1.05,1.1]))
plt.step(last.M,last.p[0]/2,label='Distribuição Teórica')
plt.hist(carlo(N,2**20,1.05)-2,bins=last.M,density=True,label='Histograma')
plt.title(r'$\beta$'+' = '+str(1.05)),plt.xlabel('M'), plt.ylabel(r'$P_{M}$'), plt.legend()
plt.show()
```

3. Se em vez de considerar o modelo de Ising de campo médio considerasse o de duas dimensões como seria o algoritmo?

 (#) Pode  Escrever aqui texto // Neste algoritmo, em vez de considerar o modelo de ising de campo médio, considerei explicitamente a interação de cada spin com os seus primeiros vizinhos ${\langle i j\rangle}$ :
$$H(s)=-\sum_{\langle i j\rangle} s_{i} s_{j}$$ 
Na evolução do sistema, podemos observar o aumento de regiões de igual spin, o que se traduz numa diminuição da energia total, partindo de uma configuração de temperatura infinita (entropia máxima) até o beta considerado.

```python
# Escreva aqui codigo
def step(mat,N,beta,it):
    for k in range(it):
        i, j = nr.randint(N), nr.randint(N)
        mat[i,j] = mat[i,j] * -1
        dE = -(2*mat[i,j] * (mat[((i+1)%N),j]+mat[((i-1)%N),j]+mat[i,((j+1)%N)]+mat[i,((j-1)%N)]))
        if nr.rand() >= min(1,np.exp(-beta*dE)):
            mat[i,j] = mat[i,j] * -1
    return mat, it, np.sum(mat)/N**2

N, beta, it = 100, 20, 2
mat = nr.choice([-1, 1], [N,N])   

fig3 = plt.figure(figsize=(16, 12))
ax = fig3.add_subplot(111)
im = ax.imshow(mat)

def animate(frame):
    global mat
    global it
    mat, it, m = np.copy(step(mat, N, beta,it))
    ax.imshow(mat,'gray')
    ax.set_title('t = '+str(it)+' M = '+str(m))
    plt.draw()
    it *= 2

anim = FuncAnimation(fig3, animate, frames=20, interval = 400)
plt.rcParams["animation.html"] = "jshtml"
plt.rcParams["animation.embed_limit"] = 100
plt.close()
anim
```
