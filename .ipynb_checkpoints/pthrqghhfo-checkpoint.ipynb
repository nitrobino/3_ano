# Exercício 1
Nas aulas discutimos a solução do modelo de Ising de campo médio com N spins. Neste trabalho de casa vamos revisitar-lo numericamente. Considere o Hamiltoniano, 
\begin{equation}
H=\frac{1}{2}\sum_{i=1}^{N}\left(1-s_{i}\frac{M-s_{i}}{N-1}\right)=\frac{(N-M)(N+M)}{2(N-1)}
\end{equation}
onde a magnetização é dada por $M=\sum_{i=1}^{N}s_{i}$ e as variáveis de spin $s_{i}$ são $\{\pm1\}$.  Para este modelo a energia é função apenas da magnetização e do número de partículas. A densidade de estados do sistema é obtida usando o número de spins $s_{i}=+1$ ($N_{+}$) e o número de spins $s_{i}=-1$ ($N_{-}$): 
\begin{align}
\Omega(N_{+},N_{-})	&=\frac{N!}{N_{+}!N_{-}!}& 	N_{+}	&=\frac{N+M}{2}	& N_{-}	&=\frac{N-M}{2}
\end{align}
Com esta parametrização,
\begin{equation}
\Omega(M)=\frac{N!}{\left(\frac{N+M}{2}\right)!\left(\frac{N-M}{2}\right)!}
\end{equation}
onde $M\in\{-N,-N+2,\dots,N-2,N\}$.

1. Calcule numericamente o valor médio exacto como função da temperatura do módulo da magnetização para $\beta$ entre 0.1 e 1.5.
\begin{align}
\left\langle \frac{|M|}{N}\right\rangle 	& =\sum_{M}\frac{|M|}{N}\Omega(M)\frac{e^{-\beta E(M)}}{Z(\beta)}  \\
Z(\beta)	& =\sum_{M}\Omega(M)e^{-\beta E(M)}
\end{align}
**Sugestão**: O Cálculo da combinatórica envolve números muito grandes. O idela é usar a fórmula de Stirling,
\begin{equation}
\log(n!)=n\log(n)-n+\frac{1}{2}\log(2\pi n)
\end{equation}
e notar que o valor médio fica inalterado se deslocar a energia $-\beta E(M)\to-\beta E(M)-C$ nas exponenciais. Logo o ideal é calcular,
$$\left\langle \frac{|M|}{N}\right\rangle =\frac{\sum_{M}\frac{|M|}{N}\Omega(M)e^{-\beta E(M)-C}}{\sum_{M}\Omega(M)e^{-\beta E(M)-C}}$$
onde C é o máximo de $\ln\Omega(M)-\beta E(M)$ para cada temperatura.

Condirando o Hamiltoniano:
$H=\frac{(N-M)(N+M)}{2(N-1)}$

Onde a magnetização M é dada por $M=\sum_{i=1}s_i$ $\in \left\{-N,-N+2,...,N\right\}$

Aproximamos uma função da densidade de estados, usando a fórmula de Sterling:

$ ln(\Omega (M,N))=ln{\frac{N!}{(\frac{N+M}{2})!(\frac{N-M}{2})!}}\approx Nln(N)-ln(\frac{N-M}{2}) \cdot \frac{N-M}{2}-ln(\frac{N+M}{2}) \cdot \frac{N+M}{2}+\frac{1}{2}[ln(2\pi N)-ln(\pi (N+M))-ln(\pi (N-M))] \equiv f(N,M)$

onde, 
 $$ \Omega(N,M) e^{-\beta E(N,M) -C} \equiv e^{ln(\Omega(N,M))}e^{-\beta E(N,M) -C} \equiv e^{f(N,M)-\beta*E(N,M)-C} $$
 
Magnetização é assitótica nos limites de altas e baixas temperaturas do problema. Notar que $\beta \propto \frac{1}{T}$. O regime em que $\beta \to \infty$  corresponde a temperaturas mais altas, neste estado as magnetizações de cada átomo são aleatórias, independentes, pelo que o valor médio dos momentos magnéticos é nulo. No limite de baixas temperaturas, a configuração em que o sistema tem energia mais baixa é correspondente a uma fase paramagnética, em que os momentos magnéticos esão todos alinhados na mesma direção. Na verdade pelo principio de exclusão de Pauli os eletrões dos átomos não ocupam estados correspondentes ao mesmo nível orbital, isto reflecte-se em interações eletroestáticas-os eletrões não estão permitidos a ocuparem regiões sobrepostas no espaço- pelo que os potenciais de interação são desta natureza. Se considerarmos $K_b=1$ a representação do sistema mostra uma alteração de fase em torno do 1K, para o qual a o valor médio da energia (que depende da magnetização) varia rápidamente. 

```python
# Hamiltoniano:

def H(N,M):
    return (N-M)*(N+M)/(2*(N-1))

#função densidade de estados dada pela fórmula de Sterling:

def f(N,M):
    f = N * np.log(N) - np.log((N-M)/2) * (N-M)/2 - np.log((N+M)/2) * (N+M)/2 + 0.5*(np.log(2*pi*N)-np.log(pi*(N+M))-np.log(pi*(N-M)))
    if M==N:
        f= 0
    if M==-N:
        f= 0
    return f

#argumento da função exponencial:

def arg(N,M,beta):
    return f(N,M)-beta*H(N,M)


#calculo do valor médio da magnetização abosluta

def avMabs(N,beta):
    num=0                         #acumulador do denominador que é o valor médio do valor absoluto da magnetização para N átomos
    denom=0                        #acumulador da função partição
    Mset=np.arange(-N,N+2,2)
    dim=len(Mset)
    C=arg(N,0,beta)                     #maximo do argumento ocorre para |M|=0
    for i in range(0,dim):              #soma para todos os valores aboslutos de M possíveis, somando os spins dois a dois
        num+=np.abs(Mset[i])/N*np.exp(arg(N,Mset[i],beta)-C)
        denom+=np.exp(arg(N,Mset[i],beta)-C)
        
    return num/denom

#avMabs(100,1.3) 

# Magnetização em função do parãmetro beta para um sistema de 1000 átomos

N=1000   
beta=np.arange(0.1,1.6,0.01)
dimb=len(beta)
valmedset=np.zeros(dimb)
for i in range(0,dimb):
    valmedset[i]=avMabs(N,beta[i])
plt.plot(beta,valmedset,label='N=1000')
plt.xlabel('beta')
plt.ylabel('<|M|>/N')
plt.legend()
plt.show()

# Magnetização em função do parãmetro beta para sistemas com N entre (64,1024) átomos:

Nset=[64,128,256,512,1024]  # variando do número de átomos na grelha
valmed=np.zeros([dimb,5])
beta=np.arange(0.1,1.6,0.01)     #variação da temperatura
dimb=len(beta)
valmedset=np.zeros(dimb)
for N in Nset:
    for i in range(0,dimb):
        valmedset[i]=avMabs(N,beta[i])
    plt.plot(beta,valmedset,label='N= ' + str(N))
    plt.xlabel('beta')
    plt.ylabel('<|M|>/N')
    plt.legend()
```

(a) Represente o histograma da amostra aleatória gerada e compare com a distribuição teórica para $\beta\in\{0.1,0.5,0.75,0.9,1.0,1.05,1.1,1.2,1.5\}$.

```python
#gerador pseudo-aleatório de uma string correspondente a uma grelha de N átomos magnetizados com valores +-1:
N=1000
M=np.zeros(N)
for i in range(0,N):
    if rd.random()<0.5:
        M[i]=-1
    else:
        M[i]=1
#print(M)


# distribuição teórica para um conjunto de k magnetizações do sistema. Distribuição é construída da alínea anterior fixando o numerador para cada valor possível de |M| normalizando pela soma das distribuições de todos os valores possíveis 
def prob_M(N,beta):
    Mset=np.arange(-N,N+2,2)
    dim=len(Mset)
    probset=np.zeros(dim)
    C=arg(N,0,beta)                  
    denom=0                    
    for i in range(0,dim):
        denom+=np.exp(arg(N,Mset[i],beta)-C)        #função partição: denominador soma a energia sobre todos os valores de M  absoluto possíveis
    for i in range(0,dim):
        num=np.exp(arg(N,Mset[i],beta)-C)           #numerador calcula a energia do sistema para um valor de M abosluto
        probset[i]=num/denom
    return Mset,probset  #retorna dois arrays referentes ao conjunto de |m| e as probabilidades de cada Mi

prob_M(10,0.6)

#variação da distribuição com o parâmetro beta para um sistema com 1000 átomos
beta=[0.1,0.5,0.75,0.9,1,1.05,1.1,1.2,1.5]
for i in range(0,len(beta)):
    x,y = prob_M(1000,beta[i])  
    plt.plot(x,y,label='beta= ' + str(beta[i]))
    plt.ylabel('frequência relativa')
    plt.xlabel('M')
    plt.title('P(teórco) para N=1000')
    plt.legend()
plt.show()

#método de inversão discreto:

def inversao_discreto(p, M):
    amostra=np.zeros(M,dtype=int)
    acum = np.zeros(len(p))
    acum[:] = p
    
    for  i in range(1,len(acum)):
        acum[i] += acum[i - 1]
        
    r = rd.random_sample(M)
    for k in range(M):
        i = 0
        while(r[k] > acum[i]):
            i += 1
        amostra[k] = i
    return amostra

#histograma da amostra aleatória gerada pelo método da inversão usando a distribuição de Boltzmann.
n=1000
amostragem=np.zeros((n,len(beta)))
for i in range(0,len(beta)):
    N=1000
    x,y=prob_M(N,beta[i])  
    y=y/sum(y) #normalização das 
    amostragem[:,i]=((inversao_discreto(y,n))-N/2)*2 #amostra de valores aleatórios de M centrada em |M|=0 e expandida de forma a compensar a translação
    N,bins,patches=plt.hist(amostragem[:,i],100,density=True)
    a,b=x,y
    plt.plot(a,b/2)
    plt.ylabel('frequência relativa')
    plt.xlabel('M')
    plt.title('dist. teórica/ dist. das v.a. de M')
plt.show()
```

(b) Meça a autocorrelação da magnetização para $\beta=1.05$,
\begin{align}
corr_{M}(\tau)	& =\frac{\left\langle M_{i}M_{i+\tau}\right\rangle -\left\langle M_{i}\right\rangle ^{2}}{\left\langle M_{i}^{2}\right\rangle -\left\langle M_{i}\right\rangle ^{2}} \\
\left\langle M_{i}M_{i+\tau}\right\rangle 	& =\frac{1}{L}\sum_{i=1}^{L}M_{i}M_{i+\tau}
\end{align}


A autocorrelação da magnetização para $\beta=1.05$ é práticamente nula.


```python
def correlate(x,tau):     # x=amostra, tau=parâmetro de correlação
    soma=0
    L=len(x)
    for i in range(L-1-tau):
        soma+=x[i]*x[i+tau]
    mit=soma/(L-1)
    mi=np.sum(x)/L
    mi2=np.sum(np.power(x,2))/L
    return (mit-mi**2)/(mi2-mi)**2



tau=1000
c=np.zeros(tau)
for i in range(tau):
    c[i]=correlate(amostragem[:,5],i+1)  #para o beta=1.05 selecionamos os valores correspondentes à coluna 5 da matriz da amostra
plt.plot(c)
plt.xlabel('tau')
plt.ylabel('correlação')
plt.show()
```

Considere o passeio aleatório no espaço de magnetizações:
\begin{align}
P(M+2|M)	&=\frac{N_{-}}{N}\min\left(1,e^{-\beta\left(E(M+2)-E(M)\right)}\right)\\
P(M-2|M)	&=\frac{N_{+}}{N}\min\left(1,e^{-\beta\left(E(M-2)-E(M)\right)}\right)
\end{align}
1. Verifique que a distribuição assimptótica da magnetização é dada pela distribuição exacta da eq. [eq:exacta] para $\beta=1.05$.



 (#) Pode  Escrever aqui texto 

```python
def randomwalk(m0,tf,beta,N):
    t = 0
    magset= np.zeros(tf+1)
    magset[0]= m0
    step = [-2,2]
    while (t < tf):
        change = rd.choice(step)
        M = magset[t]
        if rd.random()< ((N+np.sign(change)*M)/2) * min(1, np.exp(-beta*(H(M+change,N)-H(M,N)))):
            magset[t+1]= M + change
        else:
            magset[t+1]= M
            
        t+=1
    return magset
N=100
beta=1.05
tempo=[100,1000,100000]
for t in tempo:
    amostra=randomwalk(0,t,beta,N)
    n,bins,c=plt.hist(amostra,int(np.max(amostra)*0.5-np.min(amostra)*0.5),density=True)
    a,b=prob_M(N,beta)
    plt.plot(a,b)
    plt.show()
```

<span style="color:red; font-weight:bold;">Nota:</span> 0
<br><div style="color:red; font-weight:bold;">Comentário:</div> No código são mal definidas as probabilidades de aceitação e de proposta. dado que as probabilidades de saltar para magnetizações vizinhas são descritas por $$
\begin{aligned} P(M+2 | M) &=\frac{N_{-}}{N} \min \left(1, e^{-\beta(E(M+2)-E(M))}\right) \\ P(M-2 | M) &=\frac{N_{+}}{N} \min \left(1, e^{-\beta(E(M-2)-E(M))}\right) \end{aligned}
$$

teríamos que, para respeitar a condição de balanço detelhado e a escolha de Metropolis, tomar as partes $\frac{N_{-}}{N}$ ou $\frac{N_{+}}{N}$ como sendo as probabilidades de propor o salto e o $\min \left(1, e^{-\beta(E(M+2)-E(M))}\right)$ ou $\min \left(1, e^{-\beta(E(M-2)-E(M))}\right)$ como as probabilidades de aceitar o salto. O que foi feito não respeita a condição de balanço detalhado e não resulta no algorítmo correto, pois a condição de balanço detalhado é $$
\frac{A_{i j}}{A_{j i}}=\frac{E_{j i} \tilde{P}_{i}}{E_{i j} \tilde{P}_{j}} \quad \forall i, j
$$
Assim, tendo definido $A_{M+2,M} = \frac{N_{+}}{N} \min \left(1, e^{-\beta(E(M-2)-E(M))}\right) $ e $A_{M,M-2} = \frac{N_{+}}{N} \min \left(1, e^{-\beta(E(M-2)-E(M))}\right) $ e $ E{ij} = E{ji} = 0.5$ é fácil ver que a condição não se verifica.
